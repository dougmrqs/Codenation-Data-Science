```python
train = pd.read_csv('train.csv')

test = pd.read_csv('test.csv')

test.drop(['TP_ENSINO', 'TP_DEPENDENCIA_ADM_ESC', 'Q027', 'TP_STATUS_REDACAO', 'SG_UF_RESIDENCIA', 'IN_TREINEIRO', 
           'IN_BAIXA_VISAO', 'IN_CEGUEIRA', 'IN_SURDEZ', 'IN_DISLEXIA', 'IN_DISCALCULIA', 'IN_SABATISTA', 'IN_GESTANTE',
           'IN_IDOSO', 'CO_UF_RESIDENCIA', 'CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT'],
          axis=1, inplace=True)

test.fillna(0, inplace=True)
```
